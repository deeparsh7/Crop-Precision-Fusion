```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pickle as pkl
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Crop_recommendation.csv')
```

```python
# Initial explorations
print(df.head())
print(df.tail(5))
print(df.sample(5))
print("Shape of the dataset:", df.shape)
print("Columns in the dataset:", df.columns)
print("Unique labels:", df.label.unique())
print("Count of records for each label:", df.label.value_counts())
print("Unique values for each column:", df.nunique())
print("Null values in each column:", df.isnull().sum())
print("Data type of each column:")
df.info()
print(df.describe())
```

```python
# Visualizing data distributions
df1 = df.columns[:-1]
```

```python
# Boxplots for each feature
for i in df1:
    plt.figure(figsize=(15, 6))
    sns.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
# Bar plots for each feature against the label
for i in df1:
    plt.figure(figsize=(15, 6))
    sns.barplot(y=i, x='label', data=df, palette='hls')
    plt.xticks(rotation=90)
    plt.show()
```

```python

# Correlation heatmap
correlation_matrix = df.drop('label', axis=1).corr()
plt.figure(figsize=(30, 35))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=.3, center=0, square=True, linewidths=.5, annot=True)
plt.show()
```
